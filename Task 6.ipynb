```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv('Iris.csv')
```

```python
df.head()
```

```python
Target=df.Species.copy()
```

```python
Target.unique()
```

<h4>Inference- As the target variables are categorical in nature hence converting it to numerical values as this is a multiclass classification problem </h4>

```python
df['Species'].replace({'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2},inplace=True)
```

```python
df['Species']
```

<h4>Model building</h4> 

```python
X=df.drop('Species',axis=1)
y=df.Species
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(X,y)
```

<h4>Visualizing the tree</h4> 

```python
fig = plt.figure(figsize=(15,10))
_ = tree.plot_tree(model, 
                   feature_names=list(X.columns),  
                   class_names=list(Target.unique()),
                   filled=True)
```

We can now feed any new/test data to this classifer and it would be able to predict the right class accordingly.
